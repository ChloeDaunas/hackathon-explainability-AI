```python
import pandas as pd

# Load the dataset
df = pd.read_csv('HRDataset_v14.csv')

# Show the first 5 rows to identify sensitive columns
df.head()
```

```python
# 1. List the columns we found and want to remove for privacy
pii_to_remove = ['Employee_Name', 'EmpID', 'Zip', 'DOB']

# 2. Create the secure version
df_secure = df.drop(columns=pii_to_remove)

# 3. Double check the result - you should see NO names or birthdays here
print("Sanitization complete!")
print("Columns now in dataset:", df_secure.columns.tolist())
df_secure.head()
```

```python
# Save to the Colab sidebar
df_secure.to_csv('HR_Dataset_Sanitized.csv', index=False)

# Download it to your computer so you can send it to your team
from google.colab import files
files.download('HR_Dataset_Sanitized.csv')
```
